```python
import os
from subprocess import Popen, PIPE as P


class Graph:
    def __init__(self, edges, order, names):
        self.order = order
        self.edges = edges
        self.names = names
    
    def dottext(self):
        result = ["node [shape=plaintext]"] 
        result += [
            '  {0} [label=<<table>'
            '<tr><td><img src="{0:03d}.png"></img></td></tr>'
            '<tr><td> <FONT POINT-SIZE="26">{0}</FONT></td></tr>'
            '<tr><td> <FONT POINT-SIZE="26">{1}</FONT></td></tr>'
            '</table>>]'.format(num, self.names[num])

            for num in order
        ]
        result += [
            "  {} -> {} [ penwidth = 10];".format(node, edge)
            for node, elements in self.edges.items()
            for edge in elements
        ]
        return "digraph G {{\n  ranksep=0.25;\n{}\n}}".format("\n".join(result))

    def dot(self, format="png"):  # ToDo: Tratar erro
        kwargs = {} if os.name != 'nt' else {"creationflags": 0x08000000}
        p = Popen(['dot', '-T', format], stdout=P, stdin=P, stderr=P, **kwargs)
        return p.communicate(self.dottext().encode('utf-8'))[0]
    
    def __repr__(self):
        return self.dottext()
    
    def _repr_svg_(self):
        return self.dot("svg").decode("utf-8")
    
    def _repr_png_(self):
        return self.dot("png")
    
    def save(self, name, format="png"):
        with open(name, "wb") as f:
            f.write(self.dot(format))
graph = Graph(edges, order, names)
```
